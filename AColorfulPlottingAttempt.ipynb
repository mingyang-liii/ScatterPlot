```python
import pandas as pd
import seaborn as sns
```

```python
data_url = 'https://raw.githubusercontent.com/mingyang-liii/ScatterPlot/main/Reach_Cross_Section_1.csv'
CrossSection = pd.read_csv(data_url)
print(CrossSection)
```

```python
color_dict = dict({'Sand':'orange','Vegetation':'green','Channel':'blue','Bar':'brown','Crop':'red'})
sns.scatterplot(x="Distance", y="Elevation", hue="Land Cover", data=CrossSection, palette=color_dict, legend='full').set_title('Cross Section 1');
```
